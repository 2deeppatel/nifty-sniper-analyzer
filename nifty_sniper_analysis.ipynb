```python
import yfinance as yf

nifty = yf.Ticker("^NSEI")
df = nifty.history(period="5d", interval="1d")
df.head(5)
```

```python
# Calculate percent change
df['Pct Change'] = df['Close'].pct_change() * 100

#Drop NaN (first row)
df = df.dropna()

#Dispaly
print(df[['Close', 'Pct Change']])
```

```python
#Define simple sniper trade signal
def get_signal(pct):
  if pct > 0.5:
    return "Buy CE"
  elif pct < -0.5:
    return "Buy PE"
  else:
      return "No Trade"

#Apply logic to the column
df['Signal'] = df['Pct Change'].apply(get_signal)

# Show final table
print(df[['Close', 'Pct Change', 'Signal']])

#Display
print(df[['Close', 'Pct Change', 'Signal']])
```

```python
# Simulate returns (very basic model for now)
def calculate_pnl(signal):
  if signal == "Buy CE":
    return 1.0 # Assume +1%
  elif signal == "Buy PE":
    return 1.0
  else:
    return 0.0 #No Trade

#Apply to PnL column
df['PnL(%)'] = df['Signal'].apply(calculate_pnl)

#Show final results
print(df[['Close', 'Pct Change', 'Signal', 'PnL(%)']])
```

```python
df.to_csv('nifty_sniper_analysis.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert index to datetime
df.index = pd.to_datetime(df.index)

# Plot
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Close'], label='NIFTY Close', linewidth=2)

# Mark Buy CE
ce_days = df[df['Signal'] == 'Buy CE']
plt.scatter(ce_days.index, ce_days['Close'], color='green', label='Buy CE', marker='^', s=100)

# Mark Buy PE
pe_days = df[df['Signal'] == 'Buy PE']
plt.scatter(pe_days.index, pe_days['Close'], color='red', label='Buy PE', marker='v', s=100)

# Format chart
plt.title("NIFTY Sniper Trade Signals")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
